# Supervised Learning : Simple Linear Regression

```python
#random number generally between 0 and 1 , so we magnify it by multiplying with 10
x = 50 * np.random.rand(100)
```

```python
#randn mane normal distribution
y = 5 * x + np.random.randn(100)
```

```python
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
X = x.reshape(-1,1)
X.shape
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x_fit = np.linspace(-1,11)
```

```python
#matrix format e convert kora
X_fit = x_fit.reshape(-1,1)
```

```python
y_fit = model.predict(X_fit)
```

```python
plt.scatter(x,y)
plt.plot(x_fit,y_fit)
```
